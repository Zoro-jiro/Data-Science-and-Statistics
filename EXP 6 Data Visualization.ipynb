```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

> LINE CHART

```python
plt.plot(x,y)
plt.title("Line Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

> BAR CHART

```python
plt.bar(x,y)
plt.title("Line Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

```python
x = np.arange(1, 11)                 # x values (1 to 10)
y = np.random.randint(1, 100, 10)    # random integers for y

# Create bar chart
plt.bar(x, y, color='red', edgecolor='black')

# Add labels and title
plt.title("Random Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis (Random Values)")

# Show the chart
plt.show()
```

> SCATTER PLOT

```python
plt.scatter(x,y,color='cyan',edgecolor='black',s=300)
plt.title("Scatter Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

> HISTOGRAM

```python
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, color='magenta', edgecolor='black')

# Add labels and title
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

>BOX PLOT

```python
plt.boxplot(x)
plt.show()
```
