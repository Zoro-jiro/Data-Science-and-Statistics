```python
import pandas as pd 
import numpy as np
import os
```

```python
data=pd.read_csv("heart.csv")
```

```python
x=data.drop("target", axis=1)
y=data["target"]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#splitting the data into training and testing data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
```

```python

dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred4=dt.predict(x_test)
```

```python
accuracy_score (y_test,y_pred4)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

model = DecisionTreeClassifier(max_depth=3)  # limit depth
model.fit(x, y)

plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=x.columns, class_names=['No', 'Yes'])
plt.title("Simplified Decision Tree (max_depth=3)")
plt.show()
```

## ✅ Conclusion

The Decision Tree Classifier was implemented successfully, achieving about 98.5% accuracy, showing high performance and easy interpretability.
