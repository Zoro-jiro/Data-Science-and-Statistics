```python
data=pd.read_csv("heart.csv")
```

```python
x=data.drop("target", axis=1)
y=data["target"]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#splitting the data into training and testing data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score 
```

```python
svm=svm.SVC()
svm.fit(x_train, y_train)
```

```python
y_pred=svm.predict(x_test)
```

```python
accuracy_score (y_test,y_pred)
```

```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

# Select features and target
X_vis = x_train[['age', 'chol']]
y_vis = y_train

# Train SVM
model = svm.SVC(kernel='linear')
model.fit(X_vis, y_vis)

# Plot points
plt.scatter(X_vis['age'], X_vis['chol'], c=y_vis, cmap='coolwarm', edgecolors='k')

# Create grid for decision boundary
xx, yy = np.meshgrid(np.linspace(X_vis['age'].min()-1, X_vis['age'].max()+1, 500),
                     np.linspace(X_vis['chol'].min()-1, X_vis['chol'].max()+1, 500))

Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary and margins
plt.contour(xx, yy, Z, levels=[0], colors='k', linewidths=2)
plt.contour(xx, yy, Z, levels=[-1, 1], colors='k', linestyles='dashed')

plt.title("SVM Decision Boundary: Age vs Cholesterol")
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```
